```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagen = cv2.imread('gato.jpg')
imagen2 = cv2.imread('bestia.jpg')
imagen_1 = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
imagen_2 = cv2.cvtColor(imagen2, cv2.COLOR_BGR2RGB)
```

## SUMA DE IMAGENES

```python
imagen3 = cv2.add(imagen_1, imagen_2)
plt.imshow(imagen3)
```

## RESTA DE IMAGENES

```python
imagen3 = cv2.subtract(imagen_1, imagen_2)
plt.imshow(imagen3)
```

## OPERACIONES LOGICAS

```python
imagengris = cv2.imread('gato.jpg', cv2.IMREAD_GRAYSCALE)
imagengris2 = cv2.imread('bestia.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(imagengris, cmap="gray")
plt.imshow(imagengris2, cmap="gray")
```

```python
#BINARIZANDO IMAGEN 1

umbral = 158
_ , imgBin = cv2.threshold(imagengris, umbral ,255, cv2.THRESH_BINARY)

plt.imshow(imgBin, cmap="gray")
```

```python
#BINARIZANDO IMAGEN 2

umbral = 158
_ , imgBin2 = cv2.threshold(imagengris2, umbral ,255, cv2.THRESH_BINARY)

print(imgBin2)
plt.imshow(imgBin2, cmap="gray")
```

## OPERACION LOGICA AND

```python
AND = cv2.bitwise_and(imgBin, imgBin2)
plt.imshow(AND, cmap="gray")
```

## OPERACION LOGICA OR

```python
OR = cv2.bitwise_or(imgBin, imgBin2)
plt.imshow(OR, cmap="gray")
```

## OPERACION LOGICA XOR

```python
XOR = cv2.bitwise_xor(imgBin, imgBin2)
plt.imshow(XOR, cmap="gray")
```
