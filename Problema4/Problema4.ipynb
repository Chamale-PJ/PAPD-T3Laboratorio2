```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('jack.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

## Transformacion Lineal

```python
x_0= np.linspace(0,255, num=256, dtype=np.float32)
y_0=x_0

lin = np.zeros(img.shape, img.dtype)

c=2
y_0 = c * x_0
maxi_0 = np.amax(y_0)
y_0 = y_0/maxi_0 *255

lin = c * img
maxi = np.amax(lin)
lin = np.uint8(lin/maxi * 255)

fig, ax = plt.subplots(1,2)
fig.set_size_inches(8,6)

ax[0].plot(x_0,y_0)
ax[1].imshow(lin, cmap='gray')
ax[0].set_title('Funcion Lineal')
ax[1].set_title('Transformacion Lineal')
```

## Transformacion Logaritmica

```python
x_1 = np.linspace(0,255, num=256, dtype=np.float32)
loga = np.zeros(img.shape, img.dtype)

c=2
y_1 = c * np.log(1+x_1)
maxi_1 = np.amax(y_1)
y_1=y_1/maxi_1 *255

loga = c*np.log(1+img)
maxi = np.amax(loga)
loga = np.uint8(loga/maxi * 255)

fig, ax = plt.subplots(1,2)
fig.set_size_inches(8,6)

ax[0].plot(x_1,y_1)
ax[1].imshow(loga, cmap='gray')
ax[0].set_title('Funcion Logaritmica')
ax[1].set_title('Transformacion Logaritmica')
```

## Transformacion Exponencial

```python
x_2 = np.linspace(0,255, num=256, dtype=np.float32)
pot = np.zeros(img.shape, img.dtype)

g=1.9
c=1
y_2 = c * np.power(x_2, g)
maxi_2 = np.amax(y_2)
y_2 = y_2/maxi_2 * 255

pot = c*np.power(img, g)
maxi = np.amax(pot)
pot = np.uint8(pot/maxi * 255)

fig, ax = plt.subplots(1,2)
fig.set_size_inches(8,6)

ax[0].plot(x_2,y_2)
ax[1].imshow(pot, cmap='gray')
ax[0].set_title('Funcion Exponencial')
ax[1].set_title('Transformacion Exponencial')
```

## Transformacion Negativa

```python
x_3 = np.linspace(0,255, num=256, dtype=np.float32)
negativo = np.zeros(img.shape, img.dtype)

y_3 = 255 - x_3
maxi_3 = np.amax(y_3)
y_3 = y_3/maxi_3 * 255

negativo = 255 - img

fig, ax = plt.subplots(1,2)
fig.set_size_inches(8,6)

ax[0].plot(x_3,y_3)
ax[1].imshow(negativo, cmap='gray')
ax[0].set_title('Funcion Negativa')
ax[1].set_title('Transformacion Negativa')
```
