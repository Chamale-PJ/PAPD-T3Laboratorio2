```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagen = cv2.imread("gato.png")
imagen2 = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

#Pasando a Escala de grises
imagen2[:,:,1] = imagen2[:,:,0]
imagen2[:,:,2] = imagen2[:,:,0]

plt.imshow(imagen2)
```

```python
umbral = 158
_ , imgBin = cv2.threshold(imagen2, umbral ,255, cv2.THRESH_BINARY)

plt.imshow(imgBin)
```
