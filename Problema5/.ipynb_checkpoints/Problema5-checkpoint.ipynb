```python
import cv2
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Estableciendo la funcion para crear un KERNEL DE N x N

def creacion_kernel():
    print("Estableciendo el tamaño del Kernel que utilizará para hacer el procesamiento N x N")
    n = int(input("Ingrese el numero de fila: \n"))
    m = int(input("Ingrese el numero de columna: \n"))
    kernel = []
    print("************************************************************************************************")
    print("TOMA EN CUENTA QUE LOS VALORES A INGRESAR PODRIAN CAMBIAR EL TIPO DE FILTRO QUE QUIERES APLICAR") 
    print("************************************************************************************************")
    for i in range(0,n):
        fila = []
        for j in range(0,m):
            fila.append(float(input("Ingrese el valor del kernel en la posicion [" + str(i)+ "] y [" + str(j) + "] :")))
        kernel.append(fila)
        
    return np.array(kernel)

def convolutional_filter(imagen, kernel):
    alto = imagen.shape[0]
    ancho =  imagen.shape[1]
    
    altoKernel = len(kernel[:,0])
    anchoKernel = len(kernel[0,:])

    #parte entera del centro del kernel
    step = int(len(kernel[0,:])/2)

    newImg = np.zeros((alto, ancho, 1))
    
    for i in range(step, alto - step):
        newPixel = 0
        for j in range(step, ancho - step):
            ventana = imagen[i-step:i+2*step, j-step:j+2*step]
            newPixel = 0
            for m in range(0, altoKernel):
                for n in range(0, anchoKernel):
                    newPixel = newPixel + ventana[m, n] * kernel[m, n]
            newImg[i, j] = int(newPixel)
    return newImg
```

```python
imagen = cv2.imread('ruido.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# PARA ESTA PRIMERA PRUEBA SE INGRESARA UN FILTRO PROMEDIO CON UNA MATRIZ 3X3, CON VALORES DE 1/9
# EN CADA POSICION DE LA MATRIZ

a = creacion_kernel()
newImg = convolutional_filter(imagen, a)

plt.figure(figsize=(14,14))


plt.subplot(221)
plt.imshow(imagen, cmap='gray')
plt.title('Imagen con ruido')

plt.subplot(222)
plt.imshow(newImg, cmap='gray')
plt.title('Filtro Promedio')
```

```python
# PARA ESTA SEGUNDA PRUEBA SE INGRESARA UN FILTRO DE DESENFOQUE CON UNA MATRIZ 3X3, CON VALORES 1
# EN CADA POSICION DE LA MATRIZ
imagen1 = cv2.imread('prueba.jpg', cv2.IMREAD_GRAYSCALE)

a = creacion_kernel()
newImg1 = convolutional_filter(imagen1, a)

plt.figure(figsize=(14,14))


plt.subplot(221)
plt.imshow(imagen1, cmap='gray')
plt.title('Imagen original')

plt.subplot(222)
plt.imshow(newImg1, cmap='gray')
plt.title('Desenfoque')
```

```python
# PARA ESTA TERCERA PRUEBA SE UTILIZARA UNA MATRIZ DE 5X5 PARA UTILZIAR UN FILTRO DE SHARPEN
B = creacion_kernel()
newImg2 = convolutional_filter(imagen1, B)

plt.figure(figsize=(14,14))


plt.subplot(221)
plt.imshow(imagen1, cmap='gray')
plt.title('Imagen original')

plt.subplot(222)
plt.imshow(newImg2, cmap='gray')
plt.title('SHARPEN')
```

```python
# PARA ESTA SEGUNDA PRUEBA SE INGRESARA UN FILTRO SOBEL CON UNA MATRIZ 3X3, CON SUS REPECTIVOS VALORES
# EN CADA POSICION DE LA MATRIZ
imagen2 = cv2.imread('perro.jpg', cv2.IMREAD_GRAYSCALE)

c = creacion_kernel()
newImg3 = convolutional_filter(imagen2, c)

plt.figure(figsize=(14,14))


plt.subplot(221)
plt.imshow(imagen2, cmap='gray')
plt.title('Imagen original')

plt.subplot(222)
plt.imshow(newImg3, cmap='gray')
plt.title('SOBEL')
```
